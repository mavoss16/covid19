```python
import geopandas
import pandas as pd
import geopy
import numpy as np
```

```python
partners = pd.read_csv("/content/drive/Shareddrives/eatgreaterdm/personal_folders/Matthew/partners_table_ids_pre_geocode.csv")
partners.head()
#partners.dtypes
```

```python
partners["search_address"] = np.where(partners["address"].isnull(), "", partners["address"]) + ", " + np.where(partners["city"].isnull(), "", partners["city"]) + " " + np.where(partners["zip"].isnull(), "", partners["zip"])
partners.head()
partners[150:200]
```

```python
geopy.geocoders.SERVICE_TO_GEOCODER
geopy.geocoders.googlev3.GoogleV3()
```

```python
kwargs = {"api_key":""} # Enter Google Maps API key

locations = geopandas.tools.geocode(partners.search_address, provider = "google", timeout = None, **kwargs)
```

```python
locations["lon"] = locations.geometry.x
locations["lat"] = locations.geometry.y
#locations["search_address"] = locations["address"].str.replace(", USA")
#locations.rename(columns = {"address":"search_address"}, inplace = True)
locations.head()
```

```python
partners["found_address"] = locations["address"]
partners["lon"] = locations["lon"]
partners["lat"] = locations["lat"]
partners.head()
partners[150:200]
```

```python
partners.to_csv("/content/drive/Shareddrives/eatgreaterdm/personal_folders/Matthew/partners_table_ids_geocoded.csv", index = False)
```
